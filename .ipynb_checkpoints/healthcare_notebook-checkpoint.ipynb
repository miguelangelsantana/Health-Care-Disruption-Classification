### Project Methodology
COVID-19 caused major disruptions all over the world. In order to work toward normalcy, we must identify and understand pain points before building tools and resources that help vulnerable populations manage the new challenges of the world. In this analysis I will observe two datasets in order to gain a deeper understanding of how COVID-19 effected the availability of health care as well as the availability of mental health in the United States. 

#### Data and Analytical Structure
The first dataset is a collection of survey data addressing Mental Health Care (last 4 weeks) in the United States. The dataset is provided by the Centers for Disease Control and Prevention via Data.gov and can be found [here.](https://catalog.data.gov/ar/dataset/mental-health-care-in-the-last-4-weeks) The dataset was collected as part of the Household Pulse Survey. Additional information outlining the specifics of the survey can be found [here.](https://www.cdc.gov/nchs/covid19/pulse/mental-health-care.htm)

The second collection is a data summary funded by the Bill and Melinda Gates Foundation. The dataset is titled Premise General Population COVID-19 Health Services Disruption Survey 2020 and can be found via Kaggle [here.](https://www.kaggle.com/vishalsiram50/covid19-health-services-disruption-survey-2020) The project analysis will follow the OSEMN framework: Obtain, Scrub, Explore, Model and Interpret.

# Data & Packages | Obtain

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
style.use('seaborn-poster') #sets the size of the charts
style.use('ggplot')
pd.set_option('display.max_columns',50)
```

```python
mdf = pd.read_csv('gov_mentalhealth4weeks.csv')
hcdf = pd.read_csv('kaggle_healthcaredisuption.csv')
```

# Data Cleaning | Scrub
#### The main driver of this analysis is identifying which individuals 'needed' health care and did not receive it.

## Mental Health (USA) | Data.gov
Recorded values represent the percentage of applicable adults respective of each 'Indicator'

```python
# clear ID | need + received/didn't receive
mdf = mdf[mdf['Indicator'] == 'Needed Counseling or Therapy But Did Not Get It, Last 4 Weeks']
```

### Null Values | Narrow Dataset

```python
mdf = mdf[['Indicator','Group','State','Subgroup','Value']].copy()
```

```python
# mdf.isnull().sum() # dropping 47 null values
mdf.dropna(inplace=True)
```

## Gates Health Care Survey | Kaggle

```python
# clear ID | needed health care
hcdf = hcdf[hcdf['gp_post_provider_need'] == 'Yes']
```

```python
# clear ID | was health care limited? | yes/no
hcdf['gp_post_provider_visit'] = hcdf['gp_post_provider_visit'].map(lambda x: 'Yes' if x == 'Yes' else 'No')
```

### Null Values | Fill
Leveraging pre and post habits to maintain as much information as possible under the assumption that people are creatures of habit. 

```python
# filling partial null values / missed provider appointment
hcdf['gp_post_provider_why'] = hcdf['gp_post_provider_why'].fillna(hcdf['gp_pre_provider_why'])
```

```python
# filling partial null values / missed dose
hcdf['gp_post_miss_dose'] = hcdf['gp_post_miss_dose'].fillna(hcdf['gp_pre_miss_dose'])
```

## Addressing Ambiguous Values
Values in numerical columns that restrict the data type

```python
remvals = ['Not Available','Prefer not to respond','Prefer not to answer',
           'did not exist','Under 16']
cols = ['age','gender','education','gp_hh']

for col in cols:
    for val in remvals:
        hcdf = hcdf[hcdf[col]!=val]
```

```python
repvals = {'20 and over':20,'10 or more':10,'Less than once a week':1}
repcols = ['gp_hh','gp_post_num_miss_dose','gp_post_provider_num_visit',
           'gp_post_num_miss_dose']

for col in repcols:
    hcdf[col] = hcdf[col].replace(repvals)
```

## Dropping Columns

```python
hcdf = hcdf[['gender','age','geography','financial_situation','education','employment_status',
             'religion','gp_hh','gp_post_provider_visit','gp_post_provider_why',
             'gp_medication','gp_post_miss_dose','gp_post_labor_force','gp_pre_labor_force',
             'gp_pre_income','gp_post_income','country']]
```

#### Addressing Final Null Values & Target

```python
hcdf['gp_post_miss_dose'] = hcdf['gp_post_miss_dose'].fillna('No Answer')
```

```python
# renaming target
hcdf = hcdf.rename(columns={'gp_post_provider_visit':'full_care'})
```

```python
hcdf.dropna(inplace=True)
```

### Feature Conversions
#### Binary Features

```python
binvals = {'Yes':1,'No':0}

bincols = ['full_care','gp_medication','gp_post_labor_force','gp_pre_labor_force']
for col in bincols:
    hcdf[col] = hcdf[col].replace(binvals)
```

```python
gendervals = {'Female':0,'Male':1}
hcdf['gender'] = hcdf['gender'].replace(gendervals)
```

```python
trivals = {'Yes':1,'No':0,'No Answer':2}

hcdf['gp_post_miss_dose'] = hcdf['gp_post_miss_dose'].replace(trivals)
```

#### Ordinal Features

```python
educvals = {'Post graduate':18,'College or university':16,'Technical school':14,
            'Primary school':8,'Secondary/high school':12,'No formal education':0}

hcdf['education'] = hcdf['education'].replace(educvals)
```

```python
agevals = {'16 to 25 years old':25,'26 to 35 years old':35,'36 to 45 years old':45,
           'Over 45 years old':55}

hcdf['age'] = hcdf['age'].replace(agevals)
```

```python
statusvals = {'I can afford food, but nothing else':1,
              'I can afford food and regular expenses, but nothing else':2,
              'I can afford food and regular expenses, and buy new clothes once a year, but nothing else':3,
              'I cannot afford enough food for my family':4,
              'I can comfortably afford food, clothes, and furniture, and I have savings':5}

hcdf['financial_situation'] = hcdf['financial_situation'].replace(statusvals)
```

#### Type and Target Conversions

```python
hcdf['gp_hh'] = hcdf['gp_hh'].astype(int)
```

```python
hcdf.drop(['country'],axis=1,inplace=True)
```

```python
# target conversion
hcdf = hcdf.rename(columns={'gp_post_provider_visit':'full_care'})
```

## Conclusion
Our analysis is centered around dataset classifiers that negatively affect our target variable (per SHAP value analysis). Lower target variables represent limited (or no health care) received by individuals who explicitly needed health care pre/post COVID-19. 

Individuals affected by limited (or no health care) have one or more of the following qualities:
* Reason for Missing Dr. Appointment | Fear of contracting COVID-19
* Do not currently take medication
* Live in countries with lock down restrictions
* Not part of the labor force pre or post COVID-19
* Students
* Women
* Self Employed

### Limitations

Online survey responses naturally carry a bias due to the fact that certain groups of individuals are not as tech savvy as others. As a result, the representation of these groups is limited. From a cultural perspective, some individuals (depending on the country) are less willing to be honest on surveys. This may be due to regulatory influence or a general mindset passed down through generations (don't complain about things you can't change). 

## Future Work
Future work should include resource development for individuals in the 20-30 age category with a special focus on college students. This specific demographic seems especially affected with respect to limited (or no) health care.
